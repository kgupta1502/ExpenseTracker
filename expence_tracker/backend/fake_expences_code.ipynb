```python
from secrets import choice as secrets_choice
import string
import hashlib  # Added for encryption

import pandas as pd
from faker import Faker

fake = Faker()
fake.unique.clear()

def generate_password(length: int = 12) -> str:
    alphabet = string.ascii_letters + string.digits + '!@#$%^&*()-_=+'
    return ''.join(secrets_choice(alphabet) for _ in range(length))

def encrypt_password(password: str) -> str:
    """Encrypt password using SHA-256 hashing"""
    return hashlib.sha256(password.encode()).hexdigest()

records = []
for user_id in range(1, 101):
    username = fake.unique.user_name()
    email = fake.unique.email()
    password = generate_password()
    encrypted_password = encrypt_password(password)  # Encrypt the password

    records.append(
        {
            'user_id': user_id,
            'username': username,
            'email': email,
            'password': encrypted_password,  # Store encrypted password
        }
    )

df = pd.DataFrame(records)
df.to_csv('fake_users.csv', index=False)
df
```

```python
from secrets import choice as secrets_choice
import string
import hashlib  # Added for encryption
import pandas as pd
from faker import Faker

fake = Faker()
fake.unique.clear()

def generate_password(length: int = 12) -> str:
    alphabet = string.ascii_letters + string.digits + '!@#$%^&*()-_=+'
    return ''.join(secrets_choice(alphabet) for _ in range(length))

def encrypt_password(password: str) -> str:
    """Encrypt password using SHA-256 hashing"""
    return hashlib.sha256(password.encode()).hexdigest()

records = []
for user_id in range(1, 101):
    username = fake.unique.user_name()
    email = fake.unique.email()
    password = generate_password()
    encrypted_password = encrypt_password(password)  # Encrypt the password
    records.append(
        {
            'user_id': user_id,
            'username': username,
            'email': email,
            'password': password,
            'hashed pass': encrypted_password
        }
    )

df = pd.DataFrame(records)
df
```
